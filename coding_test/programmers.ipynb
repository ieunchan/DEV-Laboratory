```python
# # 대충 만든 자판
def solution(keymap, target):
    result = []
    for i in target:
        if set(i) - set(''.join(keymap)):
            return -1
        else:
            count = 0
            for j in i:
                press = min([key.index(j)+1 for key in keymap if j in key])
                count += press
            result.append(count)
    return result



print(solution(["ABACD", "BCEFD"],["ABCD","AABB"]))	# [9, 4]
print(solution(["AA"], ["B"])) # -1
```

```python
# 체육복
def solution(n, lost, reserve):

    lst, res = set(lost), set(reserve)
    stolen = lst & res
    lst -= stolen
    res -= stolen
    for i in sorted(lst):
        if i-1 in res:
            res.remove(i-1)
        elif i+1 in res:
            res.remove(i+1)
        else:
            n -= 1
    return n

solution(5,[2, 4], [1, 3, 5])	#5
solution(5,[2, 4],[3])	# 4
solution(3,[3],[1]) # 2
```

```python
# 카드 뭉치
def solution(cards1, cards2, goal):
    i, j = 0,0

    for word in goal:
        if i < len(cards1) and cards1[i] == word:
            i += 1
        elif j < len(cards2) and cards2[j] == word:
            j += 1
        return 'No'
    return 'Yes'



solution(["i", "drink", "water"],["want", "to"],["i", "want", "to", "drink", "water"]) # yes
solution(["i", "water", "drink"],["want", "to"],["i", "want", "to", "drink", "water"]) # No
```

```python
# 순위
def solution(n, results):
    win, lose = {}, {}

    for a, b in results:
        win.setdefault(a, []).append(b)
        lose.setdefault(b, []).append(a)
    def tracking(start, result):
        stack = [start]
        done = set()
        while stack:
            current = stack.pop()
            for i in result.get(current, []):
                if i not in done:
                    done.add(i)
                    stack.append(i)
        return done
    result = 0

    for player in range(1,n+1):
        win_count = len(tracking(player, win))
        lose_count = len(tracking(player,lose))
        if win_count + lose_count == n-1:
            result += 1
    return result


solution(5, [[4, 3], [4, 2], [3, 2], [1, 2], [2, 5]]) # 2
```

```python
# 1차 다트게임
def solution(dartResult):
    records, current = [],0
    while current < len(dartResult):
        if dartResult[current].isdigit():
            if dartResult[current] == '1' and current + 1 < len(dartResult) and dartResult[current + 1] == '0':
                num = 10
                current += 2
            else:
                num = int(dartResult[current])
                current += 1
            bonus = dartResult[current]
            if bonus == 'S':
                record = num ** 1
            elif bonus == 'D':
                record = num ** 2
            elif bonus == 'T':
                record = num ** 3
            current += 1
            if current < len(dartResult) and dartResult[current] in '#*':
                if dartResult[current] == '*':
                    record *= 2
                    if records:
                        records[-1] *= 2
                else:
                    record *= -1
                    current += 1
            records.append(record)
        else:
            current += 1
    return records




solution("1S2D*3T")    # 37
# solution("1D2S#10S")   # 9
# solution("1D2S0T")     # 3
# solution("1S*2T*3S")   # 23
# solution("1D#2S*3S")   # 5
# solution("1T2D3D#")    # -4
# solution("1D2S3T*")    # 59
```
