```python
def solution(numbers, hand):
    result = ''
    keypad = {
        1:(0,0), 2:(0,1), 3:(0,2),
        4:(1,0), 5:(1,1), 6:(1,2),
        7:(2,0), 8:(2,1), 9:(2,2),
        '*':(3,0), 0:(3,1), '#':(3,2)
    }
    l_position = keypad['*']
    r_position = keypad['#']

    for i in numbers:
        x, y = keypad[i]
        if i in [1,4,7]:
            result += 'L'
            l_position = (x,y)
        elif i in [3,6,9]:
            result += 'R'
            r_position = (x,y)
        else:
            l_distance = abs(l_position[0] - x) + abs(l_position[1]-y)
            r_distance = abs(r_position[0] - x) + abs(r_position[1]-y)

            if l_distance < r_distance:
                result += 'L'
                l_position = (x,y)
            elif l_distance > r_distance:
                result += 'R'
                r_position = (x,y)
            else:
                if 'l' in hand:
                    result += 'L'
                    l_position = (x,y)
                else:
                    result += 'R'
                    r_position = (x,y)
    return result

# 1 2 3
# 4 5 6
# 7 8 9
# * 0 #
solution([1, 3, 4, 5, 8, 2, 1, 4, 5, 9, 5], "right") # "LRLLLRLLRRL"
# solution([7, 0, 8, 2, 8, 3, 1, 5, 7, 6, 2], "left") # "LRLLRRLLLRR"
```
