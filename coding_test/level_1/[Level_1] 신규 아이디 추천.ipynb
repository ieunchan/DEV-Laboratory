```python
def solution(new_id):
    new_id = new_id.lower()
    result = ''
    for i in new_id:
        if i.isalnum() or i in ['-', '_', '.']:
            result += i
    while '..' in result:
        result = result.replace('..','.')
    result = result.strip('.')
    if not result:
        result += 'a'*3
    if len(result) > 15:
        result = result[:15]
        result = result.rstrip('.')
    if len(result) < 3:
        result += result[-1]* (3-len(result))
    return result



# solution("...!@BaT#*..y.abcdefghijklm") # "bat.y.abcdefghi"
# solution("z-+.^.") # "z--"
# solution("=.=") # "aaa"
# solution('abcd.eeee.')
solution('aaaaaaaaaaaaaa.a') # a..
```
