```python
def get_time(time):
    if time%100 >= 50:
        return (time//100+1)*100 + time % 10
    else:
        return time + 10

def solution(schedules, timelogs, startday):
    # timelogs에서 주말인 요소를 제거한다
    # 그냥 주말인 요소를 제거하고 그 리스트에서 schedules의 요소보다 큰게 있는지만 확인하면?
    n = len(timelogs)
    count = n
    weekday = []
    for data in range(n):
        for index, value in enumerate(timelogs[data]):
            day_index = (index + startday) % 7
            if 1 <= day_index <= 5:
                weekday.append(value)
    
    size = len(weekday) // n
    for i in range(n):
        if any(x > get_time(schedules[i]) for x in weekday[i*size:(i+1)*size]):
            count -= 1
    return count


# solution(
#     [700, 800, 1100],
#     [
#         [710, 2359, 1050, 700, 650, 631, 659], # 5 6 7 8 9 10 11
#         [800, 801, 805, 800, 759, 810, 809],
#         [1105, 1001, 1002, 600, 1059, 1001, 1100]
#     ],
#     5
# )
# # 3

solution(
    [730, 855, 700, 720],
    [
        [710, 700, 650, 735, 700, 931, 912],
        [908, 901, 805, 815, 800, 831, 835],
        [705, 701, 702, 705, 710, 710, 711],
        [707, 731, 859, 913, 934, 931, 905]
    ],
    1
)
# 2
```
