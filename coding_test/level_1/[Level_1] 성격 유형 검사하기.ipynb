```python
def solution(survey, choices):
    personality = ['A', 'N', 'C', 'F', 'M', 'J', 'R', 'T']
    mbti = {i:0 for i in personality}

    for i in range(len(survey)):
        if choices[i] == 1:
            mbti[survey[i][0]] += 3
        elif choices[i] == 2:
            mbti[survey[i][0]] += 2
        elif choices[i] == 3:
            mbti[survey[i][0]] += 1
        elif choices[i] == 4:
            pass
        elif choices[i] == 5:   
            mbti[survey[i][1]] += 1
        elif choices[i] == 6:
            mbti[survey[i][1]] += 2
        elif choices[i] == 7:
            mbti[survey[i][1]] += 3
    result = ''
    pairs = [('R', 'T'), ('C', 'F'), ('J', 'M'), ('A', 'N')]

    for a, b in pairs:
        if mbti[a] >= mbti[b]:
            result += a
        else:
            result += b

    return result

# solution(["AN", "CF", "MJ", "RT", "NA"], [5, 3, 2, 7, 5]) # "TCMA"
solution(["TR", "RT", "TR"], [7, 1, 3])	# "RCJA"
```
