```python
def solution(wallpaper):
    xx,yy = [],[]
    for y, row in enumerate(wallpaper):
        for x,value in enumerate(row):
            if value == '#':
                print(x,y)
                xx.append(x)
                yy.append(y)
    return [min(yy), min(xx), max(yy)+1, max(xx)+1]


# return lux, luy, rdx, rdy
solution([".#...", "..#..", "...#."]) # [0, 1, 3, 4]
# solution(["..........", ".....#....", "......##..", "...##.....", "....#....."]) # [1, 3, 5, 8]
```
