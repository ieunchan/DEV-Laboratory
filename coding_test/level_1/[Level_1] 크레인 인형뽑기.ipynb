```python
def solution(board, moves):
    result, count = [],0
    for row in moves:
        for x in board:
            if x[row-1]:
                if result and result[-1] == x[row-1]:
                    count += 2 # count += 1
                    x[row-1] = 0
                    result.pop()
                else:
                    result.append(x[row-1])
                    x[row-1] = 0
                break
    return count # count*2

    # 초기형: count += 1, return count * 2
    # 수정: count += 1 --> count += 2, return count
    # 수정이유: 덧셈 후 곱셈은 연산을 두번진행함. 

solution([[0,0,0,0,0],[0,0,1,0,3],[0,2,5,0,1],[4,2,4,4,2],[3,5,1,3,1]], [1,5,3,5,1,2,1,4]) # 4
```
